```python
import pandas as pd
df = pd.read_excel("E-COMMERCE DATA.xlsx")
print(df.head())
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df.duplicated().sum())
print(df['Purchase_Flag'].unique())
print(df['Wishlist_Flag'].unique())
df.columns = df.columns.str.lower()
df.to_csv("E-COMMERCE DATA.csv", index=False)

#EDA
# DISTRIBUTION ANALYSIS: Hiểu hành vi phân phối của từng features.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loyalty Score Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['user_loyalty_score'], bins=30)

plt.title('Distribution of User Loyalty Score')
plt.xlabel('User Loyalty Score')
plt.ylabel('Frequency')

plt.show()

# Affinity Score Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['user_item_affinity_score'], bins=30)

plt.title('Distribution of Affinity Score')

plt.show()

# Recommendation Score Distribution
plt.figure(figsize=(8,5))

sns.histplot(df['final_recommendation_score'], bins=30)

plt.title('Distribution of Recommendation Score')

plt.show()

# Cart Activity Outlier Check
plt.figure(figsize=(8,5))

sns.boxplot(x=df['cart_add_count'])

plt.title('Cart Add Count Outlier Analysis')

plt.show()


# CONVERSION ANALYSIS: Yếu tố nào liên quan đến Purchase
# Overall Conversion Rate
conversion_rate = df['purchase_flag'].mean() * 100

print(f'Conversion Rate: {conversion_rate:.2f}%')

# Loyalty vs Purchase: Purchasers có loyalty cao hơn không?

plt.figure(figsize=(8,5))

sns.boxplot(
    x='purchase_flag',
    y='user_loyalty_score',
    data=df
)

plt.title('Loyalty Score vs Purchase')

plt.show()

# Affinity vs Purchase
plt.figure(figsize=(8,5))

sns.boxplot(
    x='purchase_flag',
    y='user_item_affinity_score',
    data=df
)

plt.title('Affinity Score vs Purchase')

plt.show()

# Recommendation Score vs Purchase
plt.figure(figsize=(8,5))

sns.boxplot(
    x='purchase_flag',
    y='final_recommendation_score',
    data=df
)

plt.title('Recommendation Score vs Purchase')

plt.show()

# Membership Level Conversion: Membership level nào convert tốt nhất?
membership_conversion = (
    df.groupby('membership_level')['purchase_flag']
      .mean()
      .sort_values(ascending=False)
)

membership_conversion.plot(kind='bar')

plt.title('Conversion Rate by Membership Level')
plt.ylabel('Conversion Rate')

plt.show()

#ENGAGEMENT ANALYSIS: Engagement có dẫn đến purchase ko?

# Cart Activity vs Purchase
plt.figure(figsize=(8,5))

sns.boxplot(
    x='purchase_flag',
    y='cart_add_count',
    data=df
)

plt.title('Cart Activity vs Purchase')

plt.show()

# Create Loyalty Buckets: Segment users theo loyalty.
df['loyalty_bucket'] = pd.qcut(
    df['user_loyalty_score'],
    q=3,
    labels=['Low', 'Mid', 'High']
)

# Conversion by Loyalty Bucket
bucket_conversion = (
    df.groupby('loyalty_bucket')['purchase_flag']
      .mean()
)

bucket_conversion.plot(kind='bar')

plt.title('Conversion by Loyalty Segment')
plt.ylabel('Conversion Rate')

plt.show()

#CORRELATION ANALYSIS: Variables liên hệ với nhau ntn?

# Correlation Matrix
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Matrix')

plt.show()

```
